<font color=#0099ff size=4 face="黑体">将干净的分词列表存储在DataFrame空间中

```python
import jieba
import pandas as pd
import jieba.analyse
# 有的环境配置下read_table出错，因此改用read_csv
raw = pd.read_csv("金庸-射雕英雄传txt精校版.txt",
                  names = ['txt'], sep ='aaa', encoding ="utf-8" ,engine='python')

# 章节判断用变量预处理
def m_head(tmpstr):
    return tmpstr[:1]

def m_mid(tmpstr):
    return tmpstr.find("回 ")

raw['head'] = raw.txt.apply(m_head)
raw['mid'] = raw.txt.apply(m_mid)
raw['len'] = raw.txt.apply(len)
# raw['chap'] = 0

# 章节判断
chapnum = 0
for i in range(len(raw)):
    if raw['head'][i] == "第" and raw['mid'][i] > 0 and raw['len'][i] < 30 :
        chapnum += 1
    if chapnum >= 40 and raw['txt'][i] == "附录一：成吉思汗家族" :
        chapnum = 0
    raw.loc[i, 'chap'] = chapnum
    
# 删除临时变量
del raw['head']
del raw['mid']
del raw['len']


rawgrp = raw.groupby('chap')
chapter = rawgrp.agg(sum) # 只有字符串列的情况下，sum函数自动转为合并字符串
chapter = chapter[chapter.index != 0]

raws = raw[raw.chap.isin([3,4])]
raws_data=raws.iloc[list(raws.txt.apply(len)>10)]#去掉章回标题
#构建新的列表空间进行存放
df=pd.DataFrame(columns=["txt","chap"])
for m,n in enumerate(raws_data.txt):
    df.loc[m,"txt"]=n
for o,p in enumerate(raws_data.chap):
    df.loc[o,"chap"]=p
    
import jieba
import jieba.analyse
#根据细胞词库提取出相关词
jieba.load_userdict("ciku.txt")
#加入停用词库
import pandas as pd
stoplist = pd.read_csv('停用词.txt',names = ['w'], sep = 'aaa', encoding = 'utf-8')
def m_cut(intxt):
    return [w for w in jieba.cut(intxt)
        if w not in stoplist and len(w)>1]
for n in range(len(raws_data)):
    res = jieba.cut(df.txt[n]) # 精确模式
    df.loc[n,"participle"]=(' '.join(res))
df
```

<font color=#0099ff size=4 face="黑体">首先将第三章的关键词提取出来，存放在df3列表空间中,并汇集在TFre3中

```python
#构建第三章的关键词训练
df3=pd.DataFrame(columns=["txt","chap","participle"])
index=0
for n in range(len(df)):
    if df.chap[n]==3:
        df3.loc[index,"txt"]=df.txt[n]
        df3.loc[index,"chap"]=df.chap[n]
        df3.loc[index,"participle"]=df.participle[n]
        index=index+1
#添加第三章的训练集，以0.7的比重提取关键词
#其次以随机数为提取数字，将段落的关键词利用jieba提取出来
import random
import math
from sklearn.model_selection import train_test_split
c=[]
txt_3=[]
chap_3=[]
TFre3=[]
for n in range(len(df3)):
    txt_3.append(df3.txt[n])
    chap_3.append(df3.chap[n])
TF3_train, TF3_test, TF3_train_chap, TF3_test_chap=train_test_split(txt_3, chap_3, 
    test_size = 0.3, random_state = 111)
for m in TF3_train:
    TFre3+=jieba.analyse.extract_tags(m, withWeight = None)[:10]
TFre3
```

<font color=#0099ff size=4 face="黑体">将第四章的关键词提取出来，存放在df4列表空间中,并汇集在TFre4中

```python
#构建第四章的关键词训练集
df4=pd.DataFrame(columns=["txt","chap","participle"])
index=0
for n in range(len(df)):
    if df.chap[n]==4:
        df4.loc[index,"txt"]=df.txt[n]
        df4.loc[index,"chap"]=df.chap[n]
        df4.loc[index,"participle"]=df.participle[n]
        index=index+1
import random
import math
from sklearn.model_selection import train_test_split
c=[]
txt_4=[]
chap_4=[]
TFre4=[]
for n in range(len(df4)):
    txt_4.append(df4.txt[n])
    chap_4.append(df4.chap[n])
TF4_train, TF4_test, TF4_train_chap, TF4_test_chap=train_test_split(txt_4, chap_4, 
    test_size = 0.3, random_state = 111)
for m in TF4_train:
    TFre4+=jieba.analyse.extract_tags(m, withWeight = None)[:10]
TFre4
```

<font color=#0099ff size=4 face="黑体">因为事件独立，故去掉TFre3与TFre4中的相同词汇，各自放在TF3end与TF4end中

```python
#因为事件独立 固去掉相同词频
TF3end=[]
for n in TFre3:
    if n in TFre4:
        continue
    TF3end.append(n)
TF3end
```

```python
TF4end=[]
for n in TFre4:
    if n in TFre3:
        continue
    TF4end.append(n)
TF4end
```

<font color=#0099ff size=4 face="黑体">将去重后的第三章第四章关键词汇放在一起构成TFend

```python
TFend=[]
for n in TF3end:
    TFend.append(n)
for m in TF4end:
    TFend.append(m)
TFend
```

<font color=#0099ff size=4 face="黑体">按照贝叶斯公式，设计计算概率函数，判断验证词组向量属于哪类并返回该类值

```python
#计算概率函数
def concluted(sub):
#保证分子,分母不为0
    result_chap=[]
    P1=len(TF3end)/(len(TFend))
    P2=len(TF4end)/(len(TFend))
    for m in sub:
        PTF1=1
        PTF2=1
        for n in m:
            #统计在第三章里发生的概率
            ccount_3=(TF3end.count(n)+1)
            ccount_4=(TF4end.count(n)+1)
            ccount_all=(TFend.count(n)+1)
            #这一组词做乘法运算，作为本组的概率值
            PTF1*=((ccount_3/len(TF3end)*P1)/(ccount_all/(len(TFend))))
            PTF2*=((ccount_4/len(TF4end)*P2)/(ccount_all/(len(TFend))))
        #返回概率大的对应的结果
        if(PTF1>PTF2):
            result_chap.append(3)
        else:
            result_chap.append(4)
    return result_chap
```

<font color=#0099ff size=4 face="黑体">带入剩下30%的验证集TF3_test，TF4_test进行运算

```python
text_3=[]
text_4=[]
def m_cut(intxt):
    return [w for w in jieba.cut(intxt)
        if w not in stoplist and len(w)>1]
for n in TF3_test :
    text_3.append(m_cut(n))
for n in TF4_test :
    text_4.append(m_cut(n))
text_3_pre=concluted(text_3)
text_4_pre=concluted(text_4)
text_true=len(text_3_pre)*[3]+len(text_4_pre)*[4]
```

<font color=#0099ff size=4 face="黑体"># 分类报告：precision/recall/fi-score/均值/分类个数

```python
# 分类报告：precision/recall/fi-score/均值/分类个数
from sklearn.metrics import classification_report
target_names = ['page 3','page 4']
print(classification_report(text_true, text_3_pre+text_4_pre, target_names=target_names))
```

```python
import jieba
import pandas as pd
import jieba.analyse
import re
import random
# 有的环境配置下read_table出错，因此改用read_csv
raw2 = pd.read_csv("附件3_金庸-天龙八部txt精校版.txt",
                  names = ['txt'], sep ='\n', encoding ="utf-8" ,engine='python')
# 章节判断用变量预处理
def m_head_2(tmpstr):
    if re.match('.*\s.*', tmpstr) != None:
        return tmpstr[:1]


def m_mid_2(tmpstr):
    A=['一','二','三','四','五','六','七','八','九','十']
    if tmpstr in A:
        return 1
    else:
        return -1

raw2['head'] = raw2.txt.apply(m_head_2)
raw2['mid'] = raw2.txt.apply(m_mid_2)
raw2['len'] = raw2.txt.apply(len)
raw2['chap'] = 0

# 章节判断
chapnum = 0
for i in range(len(raw2)):
    if m_mid_2(raw2['head'][i])>0 and raw2['len'][i] < 20 :
        if len(raw2[raw2.chap == chapnum ]) > 2:
            chapnum += 1
    if chapnum >= 50 and raw2['txt'][i] == "附录：陈世骧先生书函" :
            chapnum = 0
    raw2.loc[i, 'chap'] = chapnum
# 删除临时变量
del raw2['head']
del raw2['mid']
del raw2['len']

rawgrp = raw2.groupby('chap')
chapter_1 = rawgrp.agg(sum) # 只有字符串列的情况下，sum函数自动转为合并字符串
chapter_1 = chapter_1[chapter_1.index != 0]
#将chapter重复的章节去掉，构成新列表空间
chapter_new=pd.DataFrame(columns=['txt','chap'])
index=0
for n in chapter_1.txt:
    if(len(n))<30:
        continue
    else:
        index+=1
        chapter_new.loc[index,'txt']=n
        chapter_new.loc[index,'chap']=index
chapter_new
```
